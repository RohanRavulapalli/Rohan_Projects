## Question 1

In order to get a broad picture of the behavior of Opioid death rates overtime, we can answer the question: **1. What has been the overall historical trend of deaths by abused Opioid usage? Has there been any recent prominent increase or decrease in the United States?**

To answer this question, a line graph showcasing the overall trend of Opioid death rates overtime can be implemented. Line charts are ideal for recording and displaying trends overtime, as they map each data point of a data set to a particular point in time, eventually making it possible to form an overall historical trendline of the data.

The "Total" property embedded within the "Any", and "Rate" dictonary keys will be plotted onto the line visualization, as this property, as previously mentioned, returns the rate of overdose deaths per 100,000 people caused by any Opioid drug. Since the aim of this question is to get a broad picture of the behavior of Opioid death rates, using the values where any form of opioid drugs are factored, seems ideal.

## Question 1 Code & Explanation

- The code down below consists of importing the report of all Opioid death rates, and then initializing the list. We then use a for loop to access the specific properties we desire (in this case, the "Total" property, is accessed from the "Any" dictonary, further embedded into the "Opioid" dictonary, originally accessed by the "Rate" dictonary


- The initalized empty list is then appended with the values of all the opioid death rates caused by any sort of Opioid drug. The list is appended by the accessing of the "Total" property explained previously


- Since the report features Opioid death rates from the years of 1999-2019, we label the x-axis "Years from 1999", as the line chart showcases how many years since 1999 have been recorded (i.e in the chart generated by the code dowm below, the x-axis value of "20", corresponds to 20 years after 1999, which converts to the year of 2019)



```python
import matplotlib.pyplot as plt

import opioids

opioid_deaths = opioids.get_opioid_deaths()

# First, we must import and generate the death by Opioids data set in Jupyter Notebook

Any_Opioid_Rates = []

for opioid in opioid_deaths:
    Any_Opioid_Rates.append(opioid["Rate"]["Opioid"]["Any"]["Total"])
    
# Now, we generate a list, Specifically, the list of Opioid death rates containing deaths by any official Opioid drug
# (Hence why use the "Any" dictonary key)

plt.plot(Any_Opioid_Rates)

# Now we create a plot of the list of Opioid death rates

plt.xlabel('Years from 1999')

plt.ylabel('Death Rates')

plt.title('Opioid Death Rates overtime')

# Finally, we title the overall graph and the axes

plt.show()

# Now we showcase the results
```

## Question #1 Summary and Answer

Based on the generated line graph, we can see that the rate of deaths per 100,000 population, involving any category of Opioid drug, has been steadily increasing overtime. Death rates by opioid's appear to have peaked around the year of 2019 (20 years after 1999 as depicted by the chart labels and values), with a total death rate (per 100,000 individuals) just shy of a rate of 16.0. Death rates caused by any type of Opioid related drug appeared to have been the lowest around 1999 (0 years after 1999 as depicted by the Line graph), with a total death rate around 2.0.

## Question 2

In order to get a potential insight regarding what makes up the bulk of overall United States Opioid rates, we can answer the question: **2. On average, which opioid drugs appear to result in the most opiod related deaths?**

A bar chart showcasing the average death rates per each type of Opioid related drug can be implemented to answer this question. Bar charts are ideal for comparing different property values, as they map the data points of each property to a visual bar shown on the bar chart, with the size of a bar representing the numeric value of an entity. With a single bar allocated to each Opioid property value(s), we can form a comparison between the various existing Opioid drugs and their corresponding average death rates.

## Question 2 Code and Explanation

- In addition to the Opioid report and data set once again being imported and generated, 4 empty lists were initialized, intended to include the number of total caused deaths for each category of Opioid drugs.


- Another for loop was created, with the intention to append the total death data points for each list. The values were appended into each list by accessing the appropriate drug property (Heroin, Cocaine, Prescription, Synthetic) from the "Opioid" and "Number" dictonary keys. Depending on the list, a different drug property was accessed


- After appending values to each of the lists, sum and len functions were used to compute the average number of caused cadeaths for each Opioid drug. Each list was summed up and divided by their legnths, resulting in the computed average number of deaths


- Eventually, all 4 averages were then plotted onto a bar graph by creating a list including each Opioid's average death values, setting X-positions (Positions relative to the x-axis on the bar chart), labels (for the chart title, axes, and bars), ticks and then using the plot.bar function to specify the locations and data set required for plotting.


```python
import matplotlib.pyplot as plt

import opioids

opioid_deaths = opioids.get_opioid_deaths()

Heroin_deaths = []

Cocaine_deaths = []

Prescription_deaths = []

Synthetic_deaths = []

# Empty lists initialized for each type of Opioid drug

for opioid in opioid_deaths:
    Heroin_deaths.append(opioid["Number"]["Opioid"]["Heroin"])
    Cocaine_deaths.append(opioid["Number"]["Opioid"]["Cocaine"])
    Prescription_deaths.append(opioid["Number"]["Opioid"]["Prescription"])
    Synthetic_deaths.append(opioid["Number"]["Opioid"]["Synthetic"])

# for loop and appendation to access intended properties and populate the initialized empty lists

avg_Heroin_deaths = sum(Heroin_deaths)/len(Heroin_deaths)

avg_Cocaine_deaths = sum(Cocaine_deaths)/len(Cocaine_deaths)

avg_Prescription_deaths = sum(Prescription_deaths)/len(Prescription_deaths)

avg_Synthetic_deaths = sum(Synthetic_deaths)/len(Synthetic_deaths)

# Computing the average death rates for each category of Opiod Drug

Opioid_avg_deaths = [avg_Heroin_deaths, avg_Cocaine_deaths, avg_Prescription_deaths, avg_Synthetic_deaths]

# Generating a list containing each of the 4 categories of opioid drugs and their average total # of caused deaths

labels = ["Heroin", "Cocaine", "Prescription", "Synthetic"]

x_positions = [0, 1, 2, 3]

# Labels and x-positions intended to allocate each average Opioid drug rate on the bar chart

plt.bar(x_positions, Opioid_avg_deaths, align='center')

# Generating the bar chart intended to graph the average Opioid drug rates for comparison purposes.

plt.xticks(x_positions, labels)

# Allocating the ticks positions onto the chart

plt.xlabel('Opioid Drug type')

plt.ylabel('Average # of Caused Deaths')

plt.title("Comparison of Opioid Drug deaths")

# Labeling the overall chart and axes

plt.show()
```

## Question #2 Summary and Answer

Based on the generated bar chart, it appears that Prescription-related opioids causes the highest total amount of deaths, on average, with an average just shy of 12,000 caused deaths. Heroin, out of the four, appears to be the least responsible for Opioid related drug deaths, with an average of roughly 6,000 total caused deaths.
